```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
from sklearn.linear_model import LinearRegression
import numpy as np
df= pd.read_csv('employee(1).csv')
print (df)
X = df[['Age']]
y = df['Salary ($)']

model = LinearRegression()
model.fit(X, y)
print("Slope (Coefficient):", model.coef_)
print("Intercept:", model.intercept_)

# Predict salary for age 40
predicted_salary = model.predict([[40]])
print("Predicted salary at age 40:", predicted_salary)
```
